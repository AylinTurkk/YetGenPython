# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time
import random 
class Kumanda:
    def __init__(self):
        self.tv_durumu = False
        self.ses = 35
        self.kanal_listesi = ["ATV","STAR TV","FOX TV","TLC","TRT"]
        self.kanal = "FOX TV"
    
    def tv_acma(self):
        if self.tv_durumu:
            print("TV Açık.")
        else:
            print("TV Açılıyor...")
            self.tv_durumu = True

    def tv_kapatma(self):
        if self.tv_durumu:
            print("TV Kapatılıyor...")
            self.tv_durumu = False
        else:
            print("TV Kapalı.")

    def ses_ayarı(self):
        yeni_ses= input("Ses ayarı kaç olsun?")
        if yeni_ses.isdigit():
            yeni_ses = int(yeni_ses)
            if yeni_ses >=0 and yeni_ses <=100:
                self.ses = yeni_ses
                print("Ses ayarı:",yeni_ses)
            else:
                print("Geçersiz değer girdiniz. 0-100 arası bir ayar seçiniz!")
        else:
            print("Lütfen bir sayı değeri giriniz!")
    
    def kanal_ekleme(self):
        yeni_kanal= input("Eklemek istediğiniz kanal adını giriniz:")
        self.kanal_listesi.append(yeni_kanal)
        print(yeni_kanal,"kanal listesine eklendi.")
    
    def kanal_sayisi(self):
        print("Kanal sayısı:",len(self.kanal_listesi))

    def rastgele_kanal(self):
        rastgele_index =random.randint(0,len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele_index]
        print("Şu anki kanal:",self.kanal)

    def tv_bilgileri(self):
        print("TV Durumu:","Açık" if self.tv_durumu else "Kapalı")
        print("TV Ses ayarı:", self.ses)
        print("Kanal Listesi:", self.kanal_listesi)
        print("Şu anki kanal:", self.kanal)
    
    def kumandayi_calistir(self):
        print("Kumanda çalıştırılıyor...")

        while True:
            if self.tv_durumu:
                print("-" *20)
                print("1: TV Kapat\n2: Ses Ayarları\n3: Kanal Ekle\n4: Kanal Sayısı\n5: Rastgele Kanala Geç\n6: Televizyon Bilgileri\nq: Çıkış")
            else:
                print("-" *20)
                print("1: TV Aç\nq: Çıkış")
            
            secim = input("İşlem seçimi yapınız:")

            if secim == "q":
                print("Kumada kapatılıyor...")
                break
            if secim == "1" :
                if self.tv_durumu:
                    self.tv_kapatma()
                else:
                    self.tv_acma
            elif secim == "2":
                self.ses_ayarı()

            elif secim == "3":
                self.kanal_ekleme()

            elif secim =="4":
                self.kanal_sayisi()

            elif secim == "5":
                self.rastgele_kanal()

            elif secim == "6":
                self.tv_bilgileri()

            else:
                print("Geçersiz seçim yaptınız. Lütfen tekrar deneyiniz.")

            if not secim:
                continue

kumanda = Kumanda()
kumanda.tv_acma()
```
